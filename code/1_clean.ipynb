# Data Cleaning

### Imports

```python
import pandas as pd
```

```python
df_train = pd.read_csv('../datasets/train.csv')
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df_train.head()
```

```python
df_train.shape
```

```python
df_train.columns = df_train.columns.str.lower().str.replace(" ","_")
```

```python
df_train.drop(columns = ['lot_frontage','alley', 'fireplace_qu', 'garage_yr_blt', 'garage_finish', 'garage_qual', 'garage_cond', 'pool_qc', 'pool_area', 'misc_feature'], inplace = True)
```

Cleaning column names and dropping ones I'm not interested in

```python
df_train.isnull().sum()
```

```python
df_train['mas_vnr_type'].fillna('No', inplace = True)
df_train['bsmt_qual'].fillna('No', inplace = True)
df_train['bsmt_cond'].fillna('No', inplace = True)
df_train['bsmt_exposure'].fillna('No', inplace = True)
df_train['bsmtfin_type_1'].fillna('No', inplace = True)
df_train['bsmtfin_type_2'].fillna('No', inplace = True)
df_train['garage_type'].fillna('No', inplace = True)
df_train['fence'].fillna('No', inplace = True)
```

```python
df_train['mas_vnr_area'].fillna(0, inplace = True)
df_train['bsmtfin_sf_1'].fillna(0, inplace = True)
df_train['bsmtfin_sf_2'].fillna(0, inplace = True)
df_train['bsmt_unf_sf'].fillna(0, inplace = True)
df_train['total_bsmt_sf'].fillna(0, inplace = True)
df_train['bsmt_full_bath'].fillna(0, inplace = True)
df_train['bsmt_half_bath'].fillna(0, inplace = True)
df_train['garage_cars'].fillna(0, inplace = True)
df_train['garage_area'].fillna(0, inplace = True)
```

According to documentation, the null values are actually zeroes or indicate a "no" or "none" in categorical columns. Fixing them for modeling

```python
df_train.info()
```

```python
df_train.isnull().sum()
```

```python
df_train['sell_diff'] = df_train['yr_sold'] - df_train['year_built']
df_train['remod_diff'] = df_train['yr_sold'] - df_train['year_remod/add']
df_train['remod_x_sell'] = df_train['remod_diff'] * df_train['sell_diff']
```

Creating some new columns that I think might be relevant
